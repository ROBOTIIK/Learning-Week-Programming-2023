```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Deklarasi variabel plate1 dan plate2 dengan resource gambar
plate1 = cv2.imread('Resource/rusianplate1.jpg')
plate2 = cv2.imread('Resource/rusianplate2.jpg')
```

```python
## melihat gambar yang akan di proses
plt.imshow(plate1,cmap='gray')
```

```python
## melihat gambar yang akan di proses
plt.imshow(plate2,cmap='gray')
```

```python
plate_cascade = cv2.CascadeClassifier('haarcascade github/haarcascade_russian_plate_number.xml')
```

```python
def blur_plate(img):
    
    plate_img = img.copy()
    ##mencari setiap plat yang ada di citra
    plate_rect = plate_cascade.detectMultiScale(plate_img)
    
    ##melakukan looping sejumlah plate
    for (x,y,w,h) in plate_rect:
         plate_extract = plate_img[y:y+h, x:x+w]
         ##melakukan proses blurring pada plat yang terdeteksi
         plate_extract = cv2.blur(plate_extract,(10,10))
         plate_img[y:y+h, x:x+w] = plate_extract

    return plate_img
```

```python
plt.imshow(blur_plate(plate1))
```

```python
plt.imshow(cv2.cvtColor(blur_plate(plate2), cv2.COLOR_BGR2RGB))
```
