```python
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread("Resource/rusianplate1.jpg")
img2 = cv2.imread("Resource/rusianplate2.jpg")
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
csc = cv2.CascadeClassifier("haarcascade github/haarcascade_russian_plate_number.xml")
```

```python
detect = csc.detectMultiScale(img1)

for (x,y,w,h) in detect:
        cv2.rectangle(img1,(x,y),(x+w,y+h),(512,128,64),5)
plt.imshow(img1)
```

```python
detect = csc.detectMultiScale(img2)

for (x,y,w,h) in detect:
        cv2.rectangle(img2,(x,y),(x+w,y+h),(128,128,128),5)
plt.imshow(img2)
```

```python
def blur(image):
    new_img = image.copy()
    detect = csc.detectMultiScale(new_img)
    
    for (x,y,w,h) in detect:
            new_plat = new_img[y:y+h, x:x+w]
            new_plat = cv2.blur(new_plat,(10,10))
            new_img[y:y+h, x:x+w] = new_plat
    return new_img
```

```python
plt.imshow(blur(img1))
```

```python
plt.imshow(blur(img2))
```
