```python
import cv2
import matplotlib.pyplot as plt

input1 = './Resource/rusianplate1.jpg'
input2 = './Resource/rusianplate2.jpg'
cek = './haarcascade github/haarcascade_russian_plate_number.xml'

img = cv2.imread(input1)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img2 = cv2.imread(input2)
img2_gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

print("before :")
plt.imshow(img_rgb)
plt.show()
plt.imshow(img2_rgb)
plt.show()

plate_cascade = cv2.CascadeClassifier(cek)
search = plate_cascade.detectMultiScale(img_gray)
search2 = plate_cascade.detectMultiScale(img2_gray)

if len(search) != 0:
    for (x, y, width, height) in search:
        img_rgb[y: y+height, x: x + width, :] = cv2.blur(img_rgb[y: y+height, x: x + width, :], (50,50))

if len(search2) != 0:
    for (x, y, width, height) in search2:
        img2_rgb[y: y+height, x: x + width, :] = cv2.blur(img2_rgb[y: y+height, x: x + width, :], (50,50))
        
print("\n\nafter : ")        
plt.subplot(1,1,1)
plt.imshow(img_rgb)
plt.show()
plt.imshow(img2_rgb)
plt.show()
```
