```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plate1 = cv2.imread('Resource/rusianplate1.jpg')
plate2 = cv2.imread('Resource/rusianplate2.jpg')
```

```python
plate_cascade = cv2.CascadeClassifier('haarcascade github/haarcascade_russian_plate_number.xml')
```

```python
def detect_plate(img):
    plate_img = img.copy()

    plate_rect = plate_cascade.detectMultiScale(plate_img)
    
    blurred_img = cv2.blur(plate_img, (9,9), 0)

    for (x,y,w,h) in plate_rect:
        mask = cv2.rectangle(plate_img,(x,y),(x+w,y+h),(255,255,255),-1)
    plate_img = np.where(mask==(255,255,255), blurred_img, plate_img) 

    return plate_img
```

```python
result1 = detect_plate(plate1)
result2 = detect_plate(plate2)
```
