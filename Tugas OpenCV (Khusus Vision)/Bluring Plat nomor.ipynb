Sebelum memblur plat maka diperlukan library: 
- opencv2 sebagai tools utama
- numpy sebagai perhitungan pixel
- matplotlib sebagai visualisasi plot gambar

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Membuat fungsi untuk memvisualisasi gambar

```python
def plot_img(img):
    plt.imshow(img, cmap="gray")
    plt.axis('off')
    plt.style.use('seaborn-v0_8')
    plt.show()
```

Mengambil ekstrak data Haarcascade Rusian plate untuk mendeteksi plat nomor pada gambar

```python
plate_detect = cv2.CascadeClassifier('haarcascade_github/haarcascade_russian_plate_number.xml')
```

import gambar yang ingin di deteksi dan masukkan dalam variabel

```python
imag1 = cv2.imread('Resource/rusianplate1.jpg')
```

```python
imag2 = cv2.imread('Resource/rusianplate2.jpg')
```

Lakukan resizing dan pengubahan gambar 

```python
imag1 = cv2.resize(imag1, (800,400))
```

```python
image1 = cv2.cvtColor(imag1, cv2.COLOR_BGR2RGB)
```

Tampilkan gambar sebelum diblur

```python
plot_img(image1)
```

Memulai deteksi plat nomor sesuai haarcascade yang sudah dimport dengan fungsi opencv

```python
plate = plate_detect.detectMultiScale(image1, 1.05, 7)
```

memulai deteksi per pixel sesuai hasil detect dan mulai memberikan highlight dan blur gambar dalam highligt

```python
for(x, y, w, h) in plate:
    cv2.rectangle(image1, (x,y), (x+w,y+h), (0,0,0), 1)
    inrec = image1[y:y+h, x:x+w]
    inrec = cv2.GaussianBlur(inrec, (23,23), 30)
    image1[y:y+inrec.shape[0], x:x+inrec.shape[1]] = inrec
```

Tampilkan gambar yang sudah di blur

```python
plot_img(image1)
```

### Lakukan hal yang sama seperti langkah diatas
